```python
def convert_genres(s):
    L = []
    for i in ast.literal_eval(s):
        L.append(i['name'])
    return L
```

```python
def cast_convert(s):
    L = []
    j = 0;
    for i in ast.literal_eval(s):
        if j>3:
            break
        L.append(i['name'])
        j = j+1
    return L
```

```python
df['cast'] = df['cast'].apply(cast_convert)
```

```python
def crew_convert(s):
    L = []
    for i in ast.literal_eval(s):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
df['crew'] = df['crew'].apply(crew_convert)
```

```python
df['genres'] = df['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
df['cast'] = df['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
df['crew'] = df['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
df['overview'] = df['overview'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
df['tags'] = df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
new_df = df[['id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'][0]
```

```python
stopword = stopwords.words('english')
ps = PorterStemmer()

def text_process(tag):
    nonpunc = [c for c in tag if c not in string.punctuation]
    nonpunc = ''.join(nonpunc)
    nonpunc = nonpunc.split()
    
    return [ps.stem(word) for word in nonpunc if word.lower() not in stopword]
```

```python
text_process(new_df['tags'][0])
```

```python
tfidf = TfidfVectorizer(analyzer=text_process, max_features=5000)
```

```python
X = tfidf.fit_transform(new_df['tags']).toarray()
```
